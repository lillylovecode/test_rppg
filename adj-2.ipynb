```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 加載圖像
image_path = 'D:\\Yun\\test\\adjustment-3.jpg'
image = cv2.imread(image_path, 0)  # 加載灰度圖像

# 顯示原始圖像
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.xticks([]), plt.yticks([])

# 使用均值濾波器去除運動偽影
kernel_size = (5,5)  # 濾波器尺寸
blurred_image = cv2.blur(image, kernel_size)

# 顯示去除偽影後的圖像
plt.subplot(122), plt.imshow(blurred_image, cmap='gray'), plt.title('Motion Artifact Reduced')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np

def apply_filters(image_path):
    image = cv2.imread(image_path)

    # Apply averaging blur
    blurred_image = cv2.blur(image, (5, 5))

    # Apply Gaussian blur
    gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

    # Apply median blur
    median_blur = cv2.medianBlur(image, 5)

    # Create a sharpening kernel
    sharpen_kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])

    # Apply the sharpening kernel
    sharpened_image = cv2.filter2D(image, -1, sharpen_kernel)

    # Convert to grayscale for edge detection
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Canny edge detector
    edges = cv2.Canny(gray_image, 100, 200)

    # Display the results
    cv2.imshow('Original Image', image)
    cv2.imshow('Averaging Blur', blurred_image)
    cv2.imshow('Gaussian Blur', gaussian_blur)
    cv2.imshow('Median Blur', median_blur)
    cv2.imshow('Sharpened Image', sharpened_image)
    cv2.imshow('Edges', edges)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Provide the path to your image
apply_filters('D:\\Yun\\test\\adjustment-3.jpg')
```
